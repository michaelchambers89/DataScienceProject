```python
import pandas as pd
import numpy as np
from sklearn import preprocessing

df_modified_file = pd.read_csv("newDataSet.csv")
df_modified_file['Arrest_number'] = (df_modified_file[['Arrest']] == True).sum(axis=1)
le = preprocessing.LabelEncoder()
#convert the categorical columns into numeric
df_modified_file['Primary Type'] = le.fit_transform(df_modified_file['Primary Type'])
df_modified_file.head(10)
```

```python
import matplotlib.pyplot as plt
import numpy as np

ax = plt.subplot(111)
ax.set_ylabel('Arrests')

df_modified_file['Arrest_number'] = (df_modified_file[['Arrest']] == True).sum(axis=1)

Year_2004=df_modified_file[df_modified_file['Year']==2004]
Year_2004.groupby(['Month']).sum().Arrest_number.plot(label="2004")

Year_2005=df_modified_file[df_modified_file['Year']==2005]
Year_2005.groupby(['Month']).sum().Arrest_number.plot(label="2005")

Year_2006=df_modified_file[df_modified_file['Year']==2006]
Year_2006.groupby(['Month']).sum().Arrest_number.plot(label="2006")

Year_2007=df_modified_file[df_modified_file['Year']==2007]
Year_2007.groupby(['Month']).sum().Arrest_number.plot(label="2007")

Year_2008=df_modified_file[df_modified_file['Year']==2008]
Year_2008.groupby(['Month']).sum().Arrest_number.plot(label="2008")

Year_2009=df_modified_file[df_modified_file['Year']==2009]
Year_2009.groupby(['Month']).sum().Arrest_number.plot(label="2009")

Year_2010=df_modified_file[df_modified_file['Year']==2010]
Year_2010.groupby(['Month']).sum().Arrest_number.plot(label="2010")

ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
```

```python
#Year_2001-2018=df_modified_file[df_modified_file['Year']!=2019]
#Year_less_2019=df_modified_file[df_modified_file['Year']!=2019 & 2018]
Year_less_2018=df_modified_file[df_modified_file['Year']<2018]
Year_less_2018.head(100) 
#is2018=df_modified_file['Year']==2018
#is2018.head()
```

```python
########################
# decision trees training
########################


import pandas as pd
from sklearn.tree import DecisionTreeClassifier # Import Decision Tree Classifier
from sklearn.model_selection import train_test_split # Import train_test_split function
from sklearn import metrics #Import scikit-learn metrics module for accuracy calculation

feature_cols = ['Primary Type','Month', 'Year']
X = Year_less_2018[feature_cols] # Features
y = Year_less_2018.Arrest_number # Target variable
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1) # 70% training and 30% test
# Create Decision Tree classifer object
clf = DecisionTreeClassifier()

# Train Decision Tree Classifer
clf = clf.fit(X_train,y_train)

#Predict the response for test dataset
y_pred = clf.predict(X_test)
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
 
```

```python
############# K-Means Algorithm ##################
dataset = df_modified_file[['Primary Type','Month', 'Arrest']]
# from sklearn.datasets import dataset
# dataset = dataset()
# dataset.data.shape
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
import numpy as np
from sklearn.cluster import KMeans


kmeans = KMeans(n_clusters = 10, random_state = 0)
clusters = kmeans.fit_predict(dataset)
kmeans.cluster_centers_.shape
#The above output shows that K-means created 10 clusters with 3 features.

from scipy.stats import mode
labels = np.zeros_like(clusters)
for i in range(10):
   mask = (clusters == i)
   labels[mask] = mode(dataset.Arrest[mask])[0]
#The following lines of code will match the learned cluster labels with the true labels found in them −

from sklearn.metrics import accuracy_score
accuracy_score(dataset.Arrest, labels)
                       
```
